# 图像标注

>标题：plt.title()
x轴名称：plt.xlabel()
y轴名称：plt.ylabel()
设置x轴范围：plt.xlim()
设置y轴范围：plt.ylim()
设置x轴刻度：plt.xticks()
设置y轴刻度：plt.yticks()
设置曲线图例：plt.legend()


```python
plt.plot([1,2,3,4],[3,5,6,9],color='black', label='Line A')
plt.plot([1,2,3,4],[2,3,1,7],color='red', marker='o',label='Line B')
plt.title('Matplotlibe Test Plit')
plt.xlabel('Xlabel')
plt.ylabel('Ylabel')
plt.xlim(0,5)
plt.ylim(0,9)
plt.xticks([0,2,4])
plt.yticks([0,3,6,9,11])
# 图例要在plot()函数里设置label
plt.legend()
plt.show()
```

## 画布与子图

```python
# 创建画布
fig = plt.figure()
# 子图
# add_subplot：一张画布占两行两列中的第一个位置
ax1 = fig.add_subplot(2,2,1)
ax1.plot([1,2,3,4],[5,6,7,8])
# add_subplot：一张画布占两行两列中的第二个位置
ax2 = fig.add_subplot(2,2,2)
```

```python
# 创建画布
fig = plt.figure()
# 子图
ax1 = fig.add_subplot(2,1,1)
ax1.plot([1,2,3,4],[5,6,7,8])
ax2 = fig.add_subplot(2,1,2)
```

## 柱状图和饼图
> plt.psd(x,NFFT=256,pad_to,Fs) 功率谱密度图
plt.specgram(x,NFFT=256,pad_to,F) 谱图
plt.cohere(x,y,NFFT=256,Fs) x-y相关性函数
plt.scatter(x,y) 散点图
pltstep(x,y,where) 步阶图
plt.hist(x,bins,normed) 直方图
坐标图、箱型图、条形图、横向条形图、极坐标图、饼图

```python
# 柱状图
data = [32,48,21,100]
labels = ['Jan','Feb','Mar',"Apr"]

plt.bar(np.arange(len(data)), data, width=0.6)
plt.xticks(np.arange(len(data)), labels)
plt.show()
```

```python
# 饼图
plt.pie([10,20,30,40], labels=['a','b','c','d'], autopct='%.1f%%', explode=[0,0,0.1,0])
plt.axis('equal')
```
